```python
import math

def f(x):
    return math.log(x)+x ** 2
def f_prime(x):
    return 1/x + 2*x
```

```python
def bisection(a, b):
    x=(a+b)/2
    print(f'x = {x : .10f} and f(x) = {f(x) : .25f}')
    if f(x)*f(a) < 0:
        return a, x
    if f(x)*f(b) < 0:
        return x, b
```

```python
a, b = 0.3, 0.7
for i in range(10):
    c, d = bisection(a, b)
    a, b = bisection(c, d)
```

```python
def false_position(a, b):
    x=(a*f(b)-b*f(a))/(f(b)-f(a))
    print(f'x = {x : .10f} and f(x) = {f(x) : .25f}')
    if f(x)*f(a) < 0:
        return a, x
    if f(x)*f(b) < 0:
        return x, b
```

```python
a, b = 0.3, 0.7
for i in range(10):
    c, d = false_position(a, b)
    a, b = false_position(c, d)
```

```python
def false_position_modified(a, b):
    x=(a*f(b)-b*f(a)/2)/(f(b)-f(a)/2)
    print(f'x = {x : .10f} and f(x) = {f(x) : .25f}')
    if f(x)*f(a) < 0:
        return a, x
    if f(x)*f(b) < 0:
        return x, b
```

```python
a, b = 0.3, 0.7
for i in range(10):
    c, d = false_position_modified(a, b)
    a, b = false_position_modified(c, d)
```

```python
def newton_rhapson(x):
    print(f'x = {x : .10f} and f(x) = {f(x) : .25f}')
    return a-(f(x)/f_prime(x))
```

```python
a = 0.5
for i in range(10):
    b = newton_rhapson(a)
    a = newton_rhapson(b)
```

```python
def secant(a, b):
    x=(a*f(b)-b*f(a))/(f(b)-f(a))
    print(f'x = {x : .10f} and f(x) = {f(x) : .25f}')
    return b, x
```

```python
a, b = 0.5, 0.7
for i in range(2):
    b, c = secant(a, b)
    c, a = secant(b, c)
    a, b = secant(c, a)
```
